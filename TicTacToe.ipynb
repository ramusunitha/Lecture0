```python
from IPython.display import clear_output

def board_design(board):
    clear_output()
    print(board[7]+'|'+board[8]+'|'+board[9])
    print('-'+'-'+'-'+'-'+'-')
    print(board[4]+'|'+board[5]+'|'+board[6])
    print('-'+'-'+'-'+'-'+'-')
    print(board[1]+'|'+board[2]+'|'+board[3])
    
def user_inst():
    print('')
    print('Welcome! This board is the Tic-Tac-Toe Board.')
    print('You will begin the game by providing your positions enter either X or O')
    print('')
    print('Instructions: When Player 1 gives 9 and if he/she is X, the X will be placed instead of number 9')
    print('')
    print('Hope this is clear')
    
def replace_board(board,user_choice,marker):
    board[user_choice] = marker
    board_design(board)
    
#   
test_board = ['#','1','2','3','4','5','6','7','8','9']    
game_board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
#
board_design(test_board)
user_inst()

#Start the Logic of the game
start_game = 'y'

while start_game == 'y':
    player_type = input("Are you player 'X' or 'O' ? ")
    if player_type == 'x':
        player_pos = int(input('OK, Player X! Which position do you want your X to go into?'))
        replace_board(game_board,player_pos,'X')
    
    if player_type == 'o':
        player_pos = int(input('OK, Player O! Which position do you want your O to go into?'))
        replace_board(game_board,player_pos,'O')

    if game_board[1]==game_board[2]==game_board[3]:
        if game_board[1]=='X':
            print('X Wins!')
            break
        elif game_board[1]=='O':
            print('O Wins!')
            break
            
    if game_board[4]==game_board[5]==game_board[6]:
        if game_board[4]=='X':
            print('X Wins!')
            break
        elif game_board[4]=='O':
            print('O Wins!')
            break
    
    if game_board[7]==game_board[8]==game_board[9]:
        if game_board[7]=='X':
            print('X Wins!')
            break
        elif game_board[7]=='O':
            print('O Wins!')
            break
    
    if game_board[1]==game_board[5]==game_board[9]:
        if game_board[1]=='X':
            print('X Wins!')
            break
        elif game_board[1]=='O':
            print('O Wins!')
            break
    
    if game_board[3]==game_board[5]==game_board[7]:
        if game_board[3]=='X':
            print('X Wins!')
            break
        elif game_board[3]=='O':
            print('O Wins!')
            break
    
    if game_board[1]==game_board[4]==game_board[7]:
        if game_board[1]=='X':
            print('X Wins!')
            break
        elif game_board[1]=='O':
            print('O Wins!')
            break
    
    if game_board[2]==game_board[5]==game_board[8]:
        if game_board[2]=='X':
            print('X Wins!')
            break
        elif game_board[2]=='O':
            print('O Wins!')
            break
    
    if game_board[3]==game_board[6]==game_board[9]:
        if game_board[3]=='X':
            print('X Wins!')
            break
        elif game_board[3]=='O':
            print('O Wins!')
            break
    
    start_game = input('Continue playing (y/n)?')
    
    if start_game == 'n':
        break
```
