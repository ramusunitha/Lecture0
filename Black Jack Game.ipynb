```python
import random

suits = ("Hearts","Clubs","Spades","Diamonds")
ranks = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')

values = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,
         'Jack':10,'Queen':10,'King':10,'Ace':11}

cash = [100,150,200,50,30,20,10,5,4,2,1,1000,250,800,900,60]
```

```python
class Card():
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.rank + " of " + self.suit
```

```python
class Deck():
    
    def __init__(self):
        
        self.all_cards = []
        
        for suit in suits:
            for rank in ranks:
                created_card = Card(suit,rank)
                self.all_cards.append(created_card)
    
    def shuffle(self):
        random.shuffle(self.all_cards)
    
    def deal_one(self):
        return self.all_cards.pop()
```

```python
class Player():
    
    def __init__(self,name):
        self.name = name
```

```python
def intro():
    print ('Hello!! Welcome to Blackjack!')
    print (' ')
    print ('Let me deal the cards for you and the dealer')
    print (' ')
    print ('You Human! You will go first!')
```

```python
class Bank():
    
    finamt = 0
    yamt = 0
    zamt = 0
    
    def __init__(self):
        
        for i in cash:
            self.finamt += i
        
        dealer_amt = []
        human_amt = []
        
        for x in range(8):
            dealer_amt.append(cash.pop())
            human_amt.append(cash.pop())
        
        for y in dealer_amt:
            print(y)
            self.yamt += y
        
        for z in human_amt:
            print(z)
            self.zamt += z
            
    def dealer_win(self):
        return self.yamt + self.zamt
        
    def human_win(self):
        return self.zamt + self.yamt
        
```
