```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
from openpyxl import *
import os
import xlsxwriter

def excel_create():
    
    workbook  = xlsxwriter.Workbook('/Users/sunitharamu/Documents/first_excel.xlsx')
    worksheet = workbook.add_worksheet()
    workbook.close()
    
def excel_heading_write():    
    sheet.column_dimensions['A'].width = 30
    sheet.column_dimensions['B'].width = 30
    sheet.column_dimensions['C'].width = 40
    sheet.column_dimensions['D'].width = 20
    
    sheet.cell(row=1, column=1).value = "First Name"
    sheet.cell(row=1, column=1).font = Font(bold=True)
    sheet.cell(row=1, column=2).value = "Last Name"
    sheet.cell(row=1, column=2).font = Font(bold=True)
    sheet.cell(row=1, column=3).value = "Email ID"
    sheet.cell(row=1, column=3).font = Font(bold=True)
    sheet.cell(row=1, column=4).value = "Contact Number"
    sheet.cell(row=1, column=4).font = Font(bold=True)

def focus1(event): 
    firstname_field.focus_set() 

def focus2(event): 
    lastname_field.focus_set() 

def focus3(event): 
    emailid_field.focus_set() 
    
def focus4(event):
    contactno_field.focus_set()
    
def clear(): 
    firstname_field.delete(0, END) 
    lastname_field.delete(0, END) 
    emailid_field.delete(0, END) 
    contactno_field.delete(0, END)
    
def insert(): 
      
    if (firstname_field.get() == "" and
        lastname_field.get() == "" and
        emailid_field.get() == "" and
        contactno_field.get() == ""): 
              print('Empty Input')
  
    else: 
        current_row = sheet.max_row 
        current_column = sheet.max_column 
        
        sheet.cell(row=current_row + 1, column=1).value = firstname_field.get() 
        sheet.cell(row=current_row + 1, column=2).value = lastname_field.get() 
        sheet.cell(row=current_row + 1, column=3).value = emailid_field.get() 
        sheet.cell(row=current_row + 1, column=4).value = contactno_field.get()
        
        wb.save('/Users/sunitharamu/Documents/first_excel.xlsx')
        clear()


if __name__=='__main__':
    window = Tk()
    window.title("My First Form")
    window.geometry('400x400')
    window.configure(background = "grey");
    firstname = Label(window ,text = "First Name")
    firstname.grid(row = 0,column = 0)
    lastname = Label(window ,text = "Last Name")
    lastname.grid(row = 1,column = 0)
    emailid = Label(window ,text = "Email Id")
    emailid.grid(row = 2,column = 0)
    contactno = Label(window ,text = "Contact Number")
    contactno.grid(row = 3,column = 0)
    firstname_field = Entry(window)
    firstname_field.grid(row = 0,column = 1)
    lastname_field = Entry(window)
    lastname_field.grid(row = 1,column = 1)
    emailid_field = Entry(window)
    emailid_field.grid(row = 2,column = 1)
    contactno_field = Entry(window)
    contactno_field.grid(row = 3,column = 1)        

    if os.path.exists('/Users/sunitharamu/Documents/first_excel.xlsx') == True:
        wb = load_workbook('/Users/sunitharamu/Documents/first_excel.xlsx')
        sheet = wb.active
    else:
        excel_create()
        wb = load_workbook('/Users/sunitharamu/Documents/first_excel.xlsx')
        sheet = wb.active
        excel_heading_write()
   
    btn = ttk.Button(window ,text="Submit",command = insert).grid(row=4,column=0)

    window.mainloop()
```
